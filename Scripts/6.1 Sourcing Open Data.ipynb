## 03 Data Cleaning & Wrangling

```python
# Check Null Values
df.isnull().sum()
```

```python
# Create Subset for Null Birth Year Values
birth_year_null = df[df['birth_year'].isnull() == True]
```

```python
birth_year_null.shape
```

```python
# Drop Null Birth Year Values
df_clean_01 = df[df['birth_year'].isnull() == False]
```

```python
df_clean_01.shape
```

```python
# Check Birth Year Values
df_clean_01['birth_year'].value_counts(dropna = False)
```

```python
# Create Subset for Accurate Birth Year Values
df_clean_02 = df_clean_01[df_clean_01['birth_year'] > 1933]
```

```python
df_clean_02.shape
```

```python
# Drop Unnecessary Columns
df_clean_02 = df_clean_02.drop(columns = ['trip_id'])
df_clean_02 = df_clean_02.drop(columns = ['bike_id'])
```

```python
df_clean_02.head()
```

```python
# Rename Weekday Column
df_clean_02.rename(columns = {'weekday' : 'day_of_week'}, inplace = True)
```

```python
df_clean_02.head()
```
